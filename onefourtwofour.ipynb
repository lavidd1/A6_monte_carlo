```python
import random as rand
import matplotlib.pyplot as plt
```

```python
# generate numdice different rolls of a 6-sided die
def roll(numdice):
    currentdice = []
    for i in range(numdice):
        currentdice.append(rand.randint(1,6))
    return currentdice

# get the number of 1s, 4s, 6s that appear in a roll
def count146(currdice):
    return currdice.count(1), currdice.count(4), currdice.count(6)

# Strategy 1
def strat1():
    numdice = 6
    keptdice = []
    kept1 = False
    kept4 = False
    
    while numdice != 0:
        numkept = 0
        curr = roll(numdice)
        #print(curr)
        num1, num4, num6 = count146(curr)
        
        # check for 1 or 4 in last 2 dice
        if(numdice == 2 and not(kept1) and not(kept4) and num1 == 0 and num4 == 0):
            keptdice = [0]*6
            break
        # check for 1 in last die    
        if(numdice == 1 and not(kept1) and num1 == 0):
            keptdice = [0]*6
            break
        # check for 4 in last die
        if(numdice == 1 and not(kept4) and num4 == 0):
            keptdice = [0]*6
            break
        
        # keep 1 if not already kept
        if(num1 > 0 and not(kept1)):
            keptdice.append(1)
            #print("kept 1\n")
            numkept += 1
            kept1 = True
            numdice -= 1
        # keep 4 if not already kept
        if(num4 > 0 and not(kept4)):
            keptdice.append(4)
            #print("kept 4\n")
            numkept += 1
            kept4 = True
            numdice -= 1
        # keep all 6s    
        for i in range(num6):
            # if neither the first 1 or first 4 have been kept, leave two dice
            if(numdice == 2 and not(kept1) and not(kept4)):
                break
            # if only one of the first 1 or first 4 have been kept, leave a die    
            if(numdice == 1 and (not(kept1) or not(kept4))):
                break
           
            keptdice.append(6)
            #print("kept 6\n")
            numkept += 1
            numdice -= 1 
        
        # if nothing was kept, keep highest die
        if(numkept == 0):
            keptdice.append(max(curr))
            #print("kept " + str(max(curr)) + "\n")
            numdice -= 1
    
    return keptdice
        
```

```python
# Show that there is apparent convergence for runs of 1000 games
# gets a players score
def get_score(kept):
    score = sum(kept)
    if(score > 0):
        score -= 5
    return score
# runs 1000 simualated games of strat1 vs strat2
def games1000():
    probs = []
    strat1wins = 0
    ties = 0
    totalgames = 0
    for j in range(1000):
        strat1score = get_score(strat1())
        strat2score = get_score(strat2())
        if(strat1score > strat2score):
            strat1wins += 1
        if(strat1score == strat2score):
            ties += 1
        totalgames +=1
        probs.append(strat1wins / totalgames)
    return probs

# initialize data to be plotted
runs = list(range(1,1001))
run1 = games1000()
run2 = games1000()
run3 = games1000()
run4 = games1000()
run5 = games1000()
run6 = games1000()
run7 = games1000()
run8 = games1000()
run9 = games1000()
run10 = games1000()
```

```python
plt.plot(runs, run1) # convergence plot
plt.plot(runs, run2)
plt.plot(runs, run3)
plt.plot(runs, run4)
plt.plot(runs, run5)
plt.plot(runs, run6)
plt.plot(runs, run7)
plt.plot(runs, run8)
plt.plot(runs, run9)
plt.plot(runs, run10)
plt.xlabel('Number of games')
plt.ylabel('Winning probability of strategy 1')
plt.title('Convergence plot of strategy 1 vs strategy 2')
plt.ylim([0.4,0.475])
plt.show()
    
```

```python
# Show normal distribution conclusion of CLT
# Modify code to produce 10 runs of 10,000 games and 100,000 games
def sim1vs2():
    probs1 = []
    probs2 = []
    for i in range(1000):
        strat1wins = 0
        strat2wins = 0
        totalgames = 0
        for j in range(1000):
            strat1score = get_score(strat1())
            strat2score = get_score(strat2())
            if(strat1score > strat2score):
                strat1wins += 1
            else:
                strat2wins += 1
            totalgames +=1
        probs1.append(strat1wins / totalgames)
        probs2.append(strat2wins / totalgames)
    return probs1

probs1 = sim1vs2()
```
